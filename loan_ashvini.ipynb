```python
# Dropping columns having all missing values more than 50%

loan_mod = loan.dropna(thresh=loan.shape[0]*0.5,how='all',axis=1)
```

```python
loan_mod.shape
```

```python
# Dropping columns having all un important  values
loan_mod = loan_mod.drop(['desc', 'member_id', 'id', 'url','delinq_amnt'], axis=1)

## Deleting column having all values as same. So not adding any importance in analysis
loan_mod = loan_mod.drop(['delinq_amnt', 'acc_now_delinq', 'application_type',
                                                         'policy_code', 'pymnt_plan', 'initial_list_status', 'chargeoff_within_12_mths',
                                                          'tax_liens', 'collections_12_mths_ex_med'], axis=1)
## Deleting string columns having most of as unique values
loan_mod = loan_mod.drop(['emp_title'], axis=1)

```

```python
loan_mod.shape
```

```python
# null_columns_dropped_data[null_columns_dropped_data.isnull().sum(axis=1) >= 2]
loan_mod['term_months'] = loan_mod['term'].str.rstrip('months')
del loan_mod['term']
```

```python
loan_mod['int_rate_percentage'] = loan_mod['int_rate'].str.rstrip('%')
del loan_mod['int_rate']
```

```python
loan_mod['emp_length_years'] = loan_mod['emp_length'].str.rstrip('years ')
loan_mod['emp_length_years'].replace({'10+': '10'}, inplace=True)
loan_mod['emp_length_years'][loan_mod['emp_length_years'] == '< 1'] = '0.5'
del loan_mod['emp_length']
```

```python
loan_mod['revol_util_percentage'] = loan_mod['revol_util'].str.rstrip('%')
del loan_mod['revol_util']
```

```python
loan_mod['issue_d_month'], loan_mod['issue_d_year'] = loan_mod[
    'issue_d'].str.split('-').str

loan_mod['last_credit_pull_d_month'], loan_mod['last_credit_pull_d_year'] = loan_mod[
    'last_credit_pull_d'].str.split('-').str

loan_mod['last_pymnt_d_month'], loan_mod['last_pymnt_d_year'] = loan_mod[
    'last_pymnt_d'].str.split('-').str

loan_mod['earliest_cr_line_month'], loan_mod['earliest_cr_line_year'] = loan_mod[
    'earliest_cr_line'].str.split('-').str
```

```python
loan_mod['issue_d_year'] = '20' + loan_mod['issue_d_year']
```

```python
# delte the rows with loan status = Current
loan_mod = loan_mod[loan_mod.loan_status !='Current']
```

```python
## Final dataset for Exploratory Data Analytics

df = loan_mod
```
